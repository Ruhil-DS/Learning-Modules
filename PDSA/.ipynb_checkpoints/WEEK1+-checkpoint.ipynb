## GCD

```python
# Euclid's algorithm

def gcd(m, n):
    a,b = min(m,n), max(m,n)
    if b%a==0:
        return a
    else:
        return gcd(a, a%b)
```

```python
class Point:
    def __init__(self, a=0, b=0):
        self.x = a
        self.y = b
    
    def translate(self, del_x, del_y):
        self.x += del_x
        self.y += del_y
        
    
    def __add__(self, p):
        return Point(self.x + p.x, self.y + p.y)
    
    def __str__(self):
        return ("("+str(self.x) + ", " + str(self.y)+ ")")
    
```

## PPA/GPA

Twin primes are pairs of prime numbers that differ by 2. For example (3, 5), (5, 7), and (11,13) are twin primes.
Write a function Twin_Primes(n, m) where n and m are positive integers and n < m , that returns all unique twin primes between m and n (both inclusive). The function returns a list of tuples and each tuple (a,b) represents one unique twin prime where n <= a < b <= m.


```python
def is_prime(a):
    flag = True
    if a <2:
        flag = False
    if a>1:
        lim = int(a**(0.5))
        for i in range(2, lim+1):
            if a%i == 0:
                flag = False
                break;
    return flag
        
```

```python
def Twin_Primes(n,m):
    twins = []
    for i in range(n, m+1):
        if is_prime(i):
            if is_prime(i+2) and (i+2<m+1):
                twins.append((i, i+2))
    return twins
```

```python
class Triangle:
    def __init__(self, a,b,c):
        self.a = a
        self.b = b
        self.c = c
        
    def is_valid(self):
        flag = True
        if (self.a + self.b)<self.c:
            flag = False
        if (self.a + self.c)<self.b:
            flag = False
        if (self.b + self.c)<self.a:
            flag = False
        if flag:
            return "Valid"
        else:
            return "Invalid"
        
    def Side_Classification(self):
        self.status = self.is_valid()
        if self.status == 'Valid':
            if (self.a == self.b) and (self.a==self.c) and (self.b==self.c):
                return "Equilateral"
            elif (self.a == self.b) or (self.a==self.c) or (self.b==self.c):
                return "Isoseles"
            else:
                return "Scalene"
        return "Invalid"
    
    def Angle_Classification(self):
        sides = [self.a, self.b, self.c]
        sides.sort()
        a = sides[0]
        b = sides[1]
        c = sides[2]
        
        
        self.status = self.is_valid()
        if self.status == "Valid":
            if (a**2 + b**2) > c**2:
                return "Acute"
            if (a**2 + b**2) < c**2:
                return "Obtuse"
            if (a**2 + b**2) == c**2:
                return "Right"
        return "Invalid"
        
    def Area(self):
        self.status = self.is_valid()
        if self.status == "Valid":
            s = (self.a + self.b + self.c)/2
            area = (s*(s-self.a)*(s-self.b)*(s-self.c))**(0.5)
            return area
        return "Invalid"
```

```python
def binarySearchIndexAndComparisons(L,k, comp=0):
    
    if len(L)<1:
        return (False, comp)
    if len(L)==1:
        comp +=1
        if L[0] == k:
            return (True, comp)
        else:
            return (False, comp)
    else:
        left = 0 #index
        right = len(L)-1 #index 
        mid = (left+right)//2 #index
        comp +=1
        if L[mid] == k:
            return (True, comp)
        elif k>L[mid]:
            left = mid+1
            return binarySearchIndexAndComparisons(L[left:], k, comp)
        elif k<L[mid]:
            right = mid
            return binarySearchIndexAndComparisons(L[:right], k, comp)
            
        
```

```python
def sortL1(s):
    return s[0]
def sortL2(s):
    return (s[0], -int(s[1:]))
    
def combinationSort(strList):
    L1 = sorted(strList, key=sortL1 )
    L2 = sorted(L1, key = sortL2 )
    
    return (L1,L2)
```

```python
L = ['d34', 'g54', 'd12', 'b87', 'g1', 'c65', 'g40', 'g5', 'd77']
combinationSort(L)
```

```python
# A.swap(index1, A, index2) --- swap index1, index2 in A
# A.swap(IndexA, B, IndexB) --- swap indexA, indexB between A and B

def mergeInPlace(A, B):
    for i in range(len(A)):
        if A[i]>B[0]:
            A.swap(i, B, 0)
            j = 1
            k = 0
            while B[j] < B[k]:
                B.swap(j, B, k)
                j +=1
                if k == len(B)-2:
                    break
                k +=1
    return(A,B)
```

```python
def selectionsort(L):
    n = len(L)
    if n<1:
        return(L)
    for i in range(n):
        mpos = i
        for j in range(i+1, n):
            if L[j]<L[mpos]:
                mpos = j
        (L[i], L[mpos]) = (L[mpos], L[i])
        print("2nd - ", i, L)
```

```python
selectionsort([9, 4, 5, 2, 3, 7, 6, 8, 1])
```

```python
def insertionsort(L):
    n = len(L)
    if n<1:
        return(L)
    for i in range(n-3):
        j = i
        while(j>0 and L[j] < L[j-1]):
            (L[j], L[j-1]) = (L[j-1], L[j])
            j = j-1
    return L
```

```python
l = [9,8,7,6,5,4,3,2,1]
print(len(l) - 3)
insertionsort(l)
```

```python
def sort_key(s):
    return (-s[1], s[0])

def DishPrepareOrder(L):
    count = dict()
    new = []
    for i in L:
        if i in count.keys():
            count[i] +=1
        else:
            count[i] = 1
            #new.append(i)
    for k,v in count.items():
        new.append((k,v))
    new = sorted(new, key=sort_key)
    final_list = []
    for i in new:
        final_list.append(i[0])
    return final_list
```

```python
def add(a,b):
    return (a+b)
def sub(a,b):
    return (a-b)
def multiply(a,b):
    return (a*b)
def divide(a,b):
    if b != 0:
        return a/b
    else:
        return None
def power(a,b):
    return (a**b)


def EvaluateExpression(exp):
    stack = []
    exp = exp.split(" ")
    operators = ['+', "-", "*", "/", "**"]
    for i in exp:
        if i not in operators:
            stack.append(int(i))
        else:
            if i == operators[0]:
                b = stack.pop()
                a = stack.pop()
                stack.append( add(a,b) )
            
            elif i == operators[1]:
                b = stack.pop()
                a = stack.pop()
                stack.append( sub(a,b) )
            
            elif i == operators[2]:
                b = stack.pop()
                a = stack.pop()
                stack.append( multiply(a,b) )
            
            elif i == operators[3]:
                b = stack.pop()
                a = stack.pop()
                stack.append( divide(a,b) )
            elif i == operators[4]:
                b = stack.pop()
                a = stack.pop()
                stack.append( power(a,b) )
    return stack[0]
```

```python
class Node:
    def __init__(self, v=None):
        self.value = v
        self.next = None
        return
    def isEmpty(self):
        if self.value == None:
            return True
        return False
    def append(self, v):
        if self.isEmpty():
            self.value = v
        elif self.next == None:
            self.next = Node(v)
        else:
            self.next.append(v)
        return
                
```

```python
def reverse(root):
    if root is None or root.next is None:
        return root
    rest = reverse(root.next)
    root.next.next = root
    root.next = None
    return rest
```

```python
def partition(arr, low, up):
    i = low+1
    j = up
    pivot = arr[low]
    while(i<=j):
        while(arr[i] < pivot and i<up):
            i +=1
        while (arr[j]>pivot):
            j -=1
        if(i<j):
            arr[i], arr[j] = arr[j], arr[i]
            i +=1
            j -=1
        else:
            i +=1
    arr[low] = arr[j]
    arr[j] = pivot
    return j
```

## Queue class

```python
class Queue:
    def __init__(self):
        self.queue = []
    
    def isEmpty(self):
        return len(self.queue)==0
    
    def addq(self, v):
        self.queue.append(v)
    
    def delq(self):
        if not self.isEmpty():
            top = self.queue[0]
            self.queue = self.queue[1:]
            return top
    
    def __str__(self):
        return str(self.queue)
```

```python
q = Queue()
q.addq(5)
q.addq(10)
q.addq(2)
print(q)
print(q.isEmpty())

q.delq()
print(q)
q.delq()
print(q)
q.delq()
print(q)
q.delq()
print(q)
print(q.isEmpty())
```

## BFS

```python
def BFS(amat, v):
    visited, parent = {}, {}
    rows, cols = amat.shape
    
    for i in range(rows):
        visited[i] = False
        parent[i] = -1
    
    q = Queue()
    
    visited[v] = True
    q.addq(v)
    
    while(not q.isEmpty()):
        j = q.delq()

        for k in neighbors(amat, j):
            if not visited[k]:
                visited[k] = True
                parent[k] = j
                q.addq(k)
    return (visited, parent)
    
```

BFS ENHANCED TO RECORD DISTANCE

```python
def BFS_DIST(amat, v):
    level, parent = {}, {}
    rows, cols = amat.shape
    
    for i in range(rows):
        level[i] = -1
        parent[i] = -1
    
    q = Queue()
    
    level[v] = 0
    q.addq(v)
    
    while(not q.isEmpty()):
        j = q.delq()

        for k in neighbors(amat, j):
            if level[k] == -1:
                level[k] = level[j] + 1
                parent[k] = j
                q.addq(k)
    return (level, parent)
    
```

```python
BFS_DIST(A, 7)
```

```python
# Global initialisation with AList
visited, parent = {}, {}

# We first need an initialisation of DFS
def DFS_init(Alist):
    for i in Alist.keys():
        visited[i] = False
        parent[i] = -1
    return

def DFS(Alist, v):
    visited[v] = True
    
    for k in Alist[v]:
        if visited[k] == False:
            visited[k] = True
            parent[k] = v
            DFS(Alist, k)
    return (visited, parent)
```

## Connectivity

```python
def components(Alist):
    Component = {}
    for i in Alist.keys():
        Component[i] = -1
    
    # Compid is the current component ID (CN)
    compid, seen = 0,0
    
    while seen <= max(Alist.keys()):
        # So long i have seen vertices less than (n-1)
        
        # Take the minimum of all the vertices which hasn't been seen ( where comp. is -1)
        startV = min( [i for i in Alist.keys() if Component[i] == -1] )
        
        # DFS_init ensures to set visited list to false for all vertices
        DFS_init(Alist)
        visited, parent = DFS(Alist, startV)
        
        for i in visited.keys():
            if visited[i]:
                seen +=1
                Component[i] = compid
        compid +=1
    return Component
    
    
```

## DFS tree - pre post number

```python
(visited,pre,post) = ({},{},{})

def DFSInitPrePost(AList):
    # Initialization   
    for i in AList.keys():
        visited[i] = False
        (pre[i],post[i]) = (-1,-1)
    return

def DFSListPrePost(AList,v,count):
    visited[v] = True
    pre[v] = count
    count = count+1
                      
    for k in AList[v]:
        if (not visited[k]):
            count = DFSListPrePost(AList,k,count)
    post[v] = count
    count = count+1           
                
    return(count)
```

```python
visited, pre, post = {}, {}, {}

def DFS_PP_init(Alist):
    for i in Alist.keys():
        visited[i] = False
        pre[i] = -1
        post[i] = -1
    return

count = 0

def DFS_PP(Alist, v, count):
    visited[v] = True
    pre[v] = count
    count +=1
    
    for k in Alist[v]:
        if not visited[k]:
            count = DFS_PP(Alist, k, count)
        post[v] = count
    count +=1
    return count
```

## TOPOLOGICAL SORTING

```python
def TopoSort(Amat):
    rows, cols = Amat.shape
    indegree = {}
    tp_list = []
    
    for c in range(cols):
        indegree[c] = 0
        for r in range(rows):
            if Amat[r,c] == 1:
                indegree[c] +=1
    
    for i in range(rows):
        j = min( k for k in range(cols) if indegree[k] == 0 )
        tp_list.append(j)
        indegree[j] -=1
        
        for k in range(cols):
            if Amat[j,k] == 1:
                indegree[k] -=1
    
    return tp_list
    
```

```python
visited = {}

def neighbours(amat, v):
    nbr = []
    for u in range(len(amat)):
        if amat[v][u] == 1:
            nbr.append(u)
    return nbr


def findConnectionLevel(n, Gmat, Px, Py):
    for i in range(n):
        visited[i] = -1
    
    q = []    
    q.append(Px)
    
    while len(q) != 0:
        j = q[0]
        q = q[1:]
        
        for k in neighbours(Gmat, j):
            if visited[k] == -1 :
                visited[k] = visited[j] + 1
                q.append(k)
    return (visited[Py] + 1)
```

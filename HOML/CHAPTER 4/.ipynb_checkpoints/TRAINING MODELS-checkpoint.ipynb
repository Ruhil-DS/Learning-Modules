# normal equation 

```python
x = 2*np.random.rand(100,1)
y = 4 + 3*x + np.random.randn(100,1) #added noise
```

```python
plt.plot(x, y, "bo")
plt.xlabel("$X_1$")
plt.ylabel("y")
fig = plt.gcf()
fig.set_size_inches(15,10)
plt.axis([0,2,0,15])
```

we know that the normal equation formula is

_


theta = (x-transpose * x)^(-1) * x-transpose * y

_



so lets make use of it

```python
x_b = np.c_[np.ones((100,1)), x] #adding the bias term = 1 to each instance
```

```python
x_b[:5]
```

```python
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
```

```python
theta_best
```

we expected theta0=4, and theta1 = 3, but got 4.2 and 2.77 respectively. close enough

we can now make predictions using this theta value we just calculated using normal equation

```python
x_new = np.array([[0], [2]])
x_new_b = np.c_[np.ones((2,1)), x_new]
```

```python
y_predict = x_new_b.dot(theta_best) #the formula -> y = theta . X
```

```python
y_predict
```

```python
plt.plot(x_new, y_predict, "r-")
plt.plot(x,y, "b.")
plt.axis([0,2,0,15])
fig = plt.gcf()
fig.set_size_inches(15,10)
```

# learning curve


```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curve(model, x, y):
    x_train, x_cv, y_train, y_cv = train_test_split(x, y, test_size=0.2)
    train_error, test_error = [], []
    
    for m in range(1, len(x_train)):
        model.fit(x_train[:m], y_train[:m])
        y_train_predict = model.predict(x_train[:m])
        y_cv_predict = model.predict(x_cv)
        
        train_error.append(mean_squared_error(y_train[:m], y_train_predict))
        test_error.append(mean_squared_error(y_cv, y_cv_predict))
    
    plt.plot(np.sqrt(train_error), "r-+",linewidth=2, label="train")
    plt.plot(np.sqrt(test_error), "b-",linewidth=2, label="test")
    plt.legend(loc="upper right")
    fig = plt.gcf()
    fig.set_size_inches(7,5)
    plt.xlabel("training set size")
    plt.ylabel("RMSE")
    
                           
        
```

```python
#learning curve for linear reg

lin_reg = LinearRegression()
plot_learning_curve(lin_reg, x,y)
plt.axis([0,80,0,3])
```

```python
#learning curve for polynomial regression
#First we need to create a pipeline

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

poly_pipe = Pipeline([
    ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
    #("std_scaler", StandardScaler()),
    ("lin_reg", LinearRegression())
    
])

plot_learning_curve(poly_pipe, x, y)
plt.title("POLYNOMIAL LEARNING CURVE WITHOUT SCALING")
plt.axis([0,80,0,30])
```

```python
#learning curve for polynomial regression
#First we need to create a pipeline

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

poly_pipe = Pipeline([
    ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
    ("std_scaler", StandardScaler()),
    ("lin_reg", LinearRegression())
    
])

plot_learning_curve(poly_pipe, x, y)
plt.title("POLYNOMIAL LEARNING CURVE WITH SCALING")
plt.axis([0,80,0,3])
```

# Logistic Regression

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
x = iris["data"][:,3:]
y = (iris["target"]==2).astype(np.int)
```

```python
x[:6]
```

```python
y[:5]
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x,y)
```

```python
x_new = np.linspace(0,3,1000).reshape(-1,1)
y_prob = log_reg.predict_proba(x_new)
```

```python
plt.plot(x_new, y_prob[:,1], "g-", label="iris virginica")
plt.plot(x_new, y_prob[:,0], "r--", label="not iris virginica")
```

```python

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]

plt.figure(figsize=(8, 3))
plt.plot(x[y==0], y[y==0], "bs")
plt.plot(x[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris virginica")
plt.text(decision_boundary+0.02, 0.15, "Decision  boundary", fontsize=14, color="k", ha="center")
plt.arrow(decision_boundary, 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
plt.arrow(decision_boundary, 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02])
```

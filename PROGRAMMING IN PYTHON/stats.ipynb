# que2

```python
freq = np.array([48,31,20,9,6,5,4,2,1,1,2,1])
num = np.arange(1,13)
```

```python
n = np.sum(freq)
```

```python
p_hat = n/(np.sum(num*freq))
```

```python
print(n, p_hat, sep='\n')
```

```python
N = 10000
p_hat_boot = np.zeros(N)
for i in range(N):
    xi = st.geom.rvs(p_hat, size=n)
    p_hat_boot[i] = n/np.sum(xi)
bias_boot = np.mean(p_hat_boot) - p_hat

var = np.var(p_hat_boot, ddof=1 )
print(np.sqrt(var))
print(bias_boot)
```

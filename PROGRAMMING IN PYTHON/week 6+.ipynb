```python
def factors(n):
    l = [n]
    for i in range(1, n//2+1):
        if n%i == 0:
            l.append(i)
    l = set(l)
    return l

def common_factors(a,b):
    a_factors = set(factors(a))
    b_factors = set(factors(b))
    
    common = set.intersection(a_factors, b_factors)
    
    return common 

def factors_upto(n):
    d = dict()
    
    for i in range(1, n+1):
        i_factors = factors(i)
        d[i] = i_factors
    return d
```

## reccursive sorting 

```python

# I'll write 3 functions, one will be obvious sort, the other will be
# using reccursion and one using inbuilt funciton

def mini(L):
    # verified
    
    '''We are creating a function named mini to find the minimum value'''
    x = L[0]
    for i in range(len(L)):
        if L[i]<=x:
            x=L[i]
            
    return x


def obvious_sort(L_original):
    # verified
    
    '''This is the obvious, time consuming sorting method
    We will use a loop to find the min value and append it to 
    a newly created list and then return that list after the loop'''

    L = list(L_original)
    x = []
    for i in range(len(L)):
        m = mini(L)
        x.append(m)
        L.remove(m)
    return x


def sort_recursive(L_original):
    # verified
    
    '''This program makes use of recurrsion'''
    
    
    L = list(L_original) # Deep copying the original list so that it stays intact
    
    
    if len(L) == (0 or 1): # If length of the list is equal to 0 or 1, we return the same list
        return L
    
    
    if len(L) > 0: # This is the main reccursive part of the function
        '''We find the minimum and store it in a variable, rewmove it from the
        list and then call same reccursion on the rest of the list'''
        m = mini(L) 
        L.remove(m)
        return [m] + sort_recursive(L)
        
        
```

## binary search

```python
def BinSearch(L, k):
    '''This binary search function returns True if we have the element K present
    in the given SORTED list. If it is not present there, we return False
    
    Binary search function takes 2 arguments/parameteres : 
    sorted list L,
    element K to find
    
    '''
    begin = 0                       # We set begin and end positions so that we can chop off the
                                    # list easily
    end = len(L)-1
    
    
    while (end-begin) >1:
        
        
        mid = (begin+end)//2
        
        if L[mid] == k:             # Element found, return True
            return True
        
        if L[mid] < k:
            begin = mid+1            # Chop off left
                                    # Keep right

        if L[mid] > k:              # Chop off right
            end = mid-1           # keep left
    if L[begin] == k or L[end]==k:
        return True
    return False
```

```python
def BinSearch_recursive(L, k, begin = 0, end = len(L)-1 ):
    '''
    This is the recursive version of binary search on sorted list L to find
    if element K is present or not
    '''
    if (end - begin == 0) or (end - begin == 1):
        # If only 1 or 2 elements are left, check individually if they are equal to K or not
        
        if (L[end] == k) or(L[begin] == k):
            return True
        else:
            return False

        
        
    if (end-begin)<0:
        # If end and begin start to get swapped, we return 0 
        return False
    
    
    if (end-begin)>1:
        ''' 
        This is the recursive part of the function
        '''
        mid = (end+begin)//2 # Calculate the middle element

        if L[mid] == k:       # Check if the middle element is ewqual to K or not
            return True
        if L[mid] < k:
            # K is in right half, chop off left. 
            # move the begin index to the (mid+1)th element
            begin = mid + 1
            return BinSearch_recursive(L, k, begin, end)        # Return the same function updated begin value

        if L[mid]>k:
            # K is in left half, chop off right
            # move the end index to the (mid-1)th element
            end = mid -1
            return BinSearch_recursive(L, k, begin, end)      # Return the same function with updated end value
        
```

```python
def spiral_iterative(left, right, n):
    p = 0
    loc = left # Initially, location is left
    p +=1
    loc = right # At P1, location is right
    p+=1    # Incrementing P once, outside the loop
    while p<=n:
        if p%2==0: # If P is even, left is shifted to the right
            loc = (left+right)/2
            left = loc
        if p%2 !=0: # If P is odd, right is shifted to the left
            loc = (left+right)/2
            right = loc
        p +=1 
    return loc # Returning the location as the result

        
```

```python
spiral_iterative(0, 4,4)
```

```python
def spiral_recursive(left, right, n):
    if n <0 : # If N is negative
        return "error, enter positive n"
    
    if n == 0: # If N is 0, return left as the position
        return left
    
    if n==1: # If N is 1, return right as the position
        return right
    
    if n==2: # Base case, if N is 2, return the mid point of left and right
        return (left+right)/2
    
    elif n%2==0: # If even, shift left towards right
        loc = (left+right)/2
        left = loc
    elif n%2 != 0: # If N is odd, shift right towards left
        loc = (left+right)/2
        right = loc
        
    if n>2: # Recursive part, returning same funciton with parameter changed from n to (n-1)
        return spiral_recursive(left, right, n-1)
```

```python
# sorted in non des. order
# should be asc. ordered
def non_decreasing(L):
    if len(L) == 1:
        return True
    
    L_init = L[:-1]
    last = L[-1]
    
    if len(L)>1:
        if last >= L_init[-1]:
            return non_decreasing(L_init)
        else:
            return False
```

```python
# matrix addition

def empty_mat(n, m):
    '''A funciton to create an empty matrix with all 0s'''
    '''                    VERIFIED                    '''
    C = []
    for i in range(n):
        C.append([])
        for j in range(m):
            C[i].append(0)
    return C

def mat_add(A,B):
    nA = len(A) # Rows of A
    nB = len(B) # Rows of B
    mA = len(A[0]) # Columns of A
    mB = len(B[0]) # Columns of B
    if (nA !=nB) or (mA != mB): 
        return "Addition not possible"
    
    C = empty_mat(nA, mA)
    
    for i in range(nA):
        for j in range(mA):
            C[i][j] = A[i][j] + B[i][j]
    return C
            
```

```python
def is_reachable(grid):
    n = len(grid)
    m = len(grid[0])
    for i in range(n):
        for j in range(m):
            if grid[i][j] == 'B':
                pos_b = (i,j)
            if grid[i][j] == 'G':
                pos_g = (i,j)

    if ( pos_g[0] <= pos_b[0] ) and ( pos_g[1] >= pos_b[1] ):
        x = abs(pos_g[0] - pos_b[0])
        y = abs(pos_b[1] - pos_g[1])
        
        dist = x+y
        
        return (True, dist)
    
    return (False, None)
        
```

```python
class Point():
    def __init__(self, x,y):
        self.x = x
        self.y = y
    
    def distance(self):
        dist = (x**2 + y**2)**(0.5)
        return dist
    def is_origin(self):
        if (self.x==0) and (self.y==0):
            return True
        else:
            return False
    def on_xaxis(self):
        if self.y==0:
            return True
        else:
            return False
    def on_yaxis(self):
        if self.x==0:
            return True
        else:
            return False
    def quadrant(self):
        # ['first', 'second', 'third', 'fourth']
        if (self.x >0) and (self.y >0):
            return 'first'
        if (self.x <0) and (self.y >0):
            return 'second'
        if (self.x <0) and (self.y <0):
            return 'third'
        if (self.x >0) and (self.y <0):
            return 'fourth'
    def slope(self):
        return self.y/self.x
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def print_info(self):
        tup = (self.x, self.y)
        print("(",tup[0], ",", tup[1], ")", sep="")
    def scale(self, s):
        self.x *=s
        self.y *=s
    def reflect_about_x(self):
        self.y = -self.y
    
    def reflect_about_y(self):
        self.x = -self.x
    
    def add(self, V):
        v = Vector(0,0)
        v.x = self.x + V.x
        v.y = self.y + V.y
        return v
        
```

```python
class Country:
    def __init__(self):
        self.name = None
        self.capital = None

    def set_name(self, name):
        self.name = name

    def set_capital(self, capital):
        self.capital = capital

    def display(self):
        print(f'{self.capital} is the capital of {self.name}')

```

```python
n = int(input())
countries = []
for i in range(n):
    C = Country()
    inp1 = input()
    inp2 = input()
    C.set_name(inp1)
    C.set_capital(inp2)
    countries.append(C)
    C.display()
    
```

```python
def is_prime(x):
    if x == 1 or x==0:
        return False
    for i in range(2, x//2 + 1):
        if x%i == 0:
            return False
    return True

def primes_galore(L):
    n = len(L)
    primes = []
    for i in range(n):
        if is_prime(i) is True:
            if is_prime(L[i]):
                primes.append( L[i] )
    return primes
```

```python
primes_galore([1, 3, 11, 18, 17, 23, 6, 8, 10])
```

```python
def pattern(a,b):
    #a lines, b circles in 1st and last line
    pat = []
    s = ''
    pat.append("o"*b)
    s += "o"
    s += " "*(b-2)
    s += "o"
    for i in range(a-2):
        pat.append(s)
    
    pat.append("o"*b)
    
    for i in (pat):
        print(i)
```
